```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from source.modeling import *
```

```python
df = pd.read_csv('data/ml1.csv')
```

```python
X = df.loc[:, df.columns != 'price'].to_numpy()
y = df['price'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1234, test_size=0.20)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
lin_reg_model = OLS_linear_regression(X_train, y_train)
score_train = lin_reg_model.score(X_train, y_train)
score_test = lin_reg_model.score(X_test, y_test)
print(f'train score:\t{score_train}\ntest score:\t{score_test}')
```

```python
scalings = ['std', 'robust', 'minmax']
for scaling in scalings:
    lin_reg_model = OLS_linear_regression_scaled(X_train, y_train, scaling)
    score_train = lin_reg_model.score(X_train, y_train)
    score_test = lin_reg_model.score(X_test, y_test)
    print(f'{scaling} scaling')
    print(f'\ttrain score:\t{score_train}\n\ttest score:\t{score_test}\n')
```

Let's try to validate which degree is the best (1 to 3) for a potential polynomial regression fitting using 3-fold cross validation.

```python
plot_poly_reg_validation_curves(X_train, y_train, 3, 3)
```

As we can see, the accuracy on cross-validated training sets is very poor for every degree higher than one. Supposedly not beacause the underlying target is in fact purely linear, but because the polynomially transformed feature space is now too large to search with our limited training set.
